```python
covid_data = pd.read_csv("/content/covid_test_data.csv")
covid_data
```

```python
covid_data.loc[:, 'fever severity': 'fatigue severity'] # all rows, from column 'fever severity' to column 'fatigue severity' are for training
```

```python
x_2 = covid_data.loc[:, 'fever severity': 'fatigue severity'].values
```

```python
covid_data.loc[:, 'covid result'] # results column (covid result)
```

```python
y_2 = covid_data.loc[:, 'covid result'].values
```

```python
from sklearn.model_selection import train_test_split
x_2_train, x_2_test, y_2_train, y_2_test = train_test_split(x_2, y_2, test_size=0.3) # trying to save 30% as testing data
```

```python
# now initiating the classifier

from sklearn import tree
clf = tree.DecisionTreeClassifier()

# training the classifier with x_train and y_train
clf.fit(x_2_train, y_2_train)
```

```python
# now make prediction
predictions = clf.predict(x_2_test)
print(predictions)

# actual results
print(y_2_test)
```

```python
# show accuracy level with %
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(predictions, y_2_test) * 100
print("Accuracy of Predictions: " + str(accuracy) + "%")
print()
```
